```python
import requests
import time
from bs4 import BeautifulSoup
import pandas as pd
from requests_html import HTMLSession
from fake_useragent import UserAgent
from random import randint
from time import sleep
```

```python
links = []


for i in range(1,25):
    #on crée notre url
    url = "https://www.paruvendu.fr/auto-moto/listefo/default/default?auto-typeRech=&reaf=1&r2=&px1=&md=&codeINSEE=&lo=&pa=&ray=15&r=VVO00000&r1=&trub=&nrj=&km1=&co2=&a0=&a1=&npo=0&tr=&pf0=&pf1=&fulltext=&codPro=0&p=" + str(i)
    #on fait la requete vers cet url
    response=requests.get(url)
#Si la réponse est bonne on crée notre objet soup avec beautifulsoup
    if response.ok:
        soup=BeautifulSoup(response.text)
        #on va chercher les élements qui nous interesse 
        ads = soup.findAll("div", {"class": 'lazyload_bloc ergov3-annonce ergov3-annonceauto'})

        # sur chaque div on va chercher le <a> et pour chaque a on va recuperer le lien et on l'ajoute à notre liste
        for l in ads:
            a = l.find('a')
            link = a['href']
            links.append(link)
        #on ajoute du delai entre chaque requête pour reduire le risque d'etre bloquer par le site
        time.sleep(2)
    print("page №" + str(i)+ " scrappé")
print(str(len(links)) + " annonces récupérées")
# print(len(ads))
```

```python
with open('annonces_pagination.csv' , 'w') as file:
    for link in links:
        file.write(link + '\n')
```

```python
#Le contenu qu'on cherche à afficher n'est pas affiché par le html mais le javascript ,
# beautifulsoup ne prenant pas en compte le contenu dynamique on doit importer une librairie 
# from requests_html import HTMLSession
# session = HTMLSession()



with open('annonces_pagination.csv' , 'r') as file:
    with open('annonces_auto.csv' , 'w' , encoding='utf-8') as outf:
        outf.write('marque, version , annee , kilometrage , energie , emission de CO2 , consommation mixte , transmission , nombre de portes , puissance fiscale,nombre de places , lieu, prix\n')
        for row in file:
            response = requests.get(row)
            if response.ok:
                soup = BeautifulSoup(response.content , 'html.parser')
                try:
                    marque_full = soup.find('h1').text
                    marque = marque_full.split()[0]
                    marque = marque.strip()
                except:
                    marque = "null"
                
                try:
                    version = soup.find('li' , {'class' : 'vers'}).find("span").text
                    version = version.strip()
                except:
                    version = "null"

                try:
                    date = soup.find('li' , {'class' : 'ann'}).find("span").next
                    date = date.split()[-1]
                    date = date.strip()
                except:
                    date = "null"
                
                try:
                    kil_full = soup.find('li' , {'class' : 'kil'}).find("span").text
                    kil =  kil_full.replace("km",  "")
                    kil = kil.strip()
                except:
                    kil = "null"
                
                try:
                    energie = soup.find('li' , {'class' : 'en'}).find("span").text
                    energie = energie.strip()
                except:
                    energie = "null"

                try:
                    emission_full = soup.find('li' , {'class' : 'emiss'}).find("span").text
                    emission =  emission_full.split()[0]
                    emission = emission.strip()
                except:
                    emission = "null"
                
                try:
                    cons_full  = soup.find('li' , {'class' : 'cons'}).find("span").text
                    cons = cons_full.split()[0]
                    cons = cons.strip()
                except:
                    cons = "null"

                try:
                    transmission = soup.find('li' , {'class' : 'vit'}).find("span").text
                    transmission = transmission.strip()
                except:
                    transmission = "null"

                try:    
                    porte = soup.find('li' , {'class' : 'carro'}).find("span").text
                    porte = porte.replace('portes' , '')
                    porte = porte.split()[0]
                    porte = porte.strip()
                except:
                    porte = "null"
                
                try:
                    cv = soup.find('li' , {'class' : 'puiss'}).find("span").text
                    cv = cv.replace('CV' , '')
                    cv = cv.strip()
                except:
                    cv = "null"
                
                try:
                    place = soup.find('li' , {'class' : 'por'}).find("span").text
                    place = place.split()[0]
                    place = place.strip()
                except:
                    place = "null"
                
                try:
                    lieu_full = soup.find('div' , {'class' : 'opt19_listlinks'}).find("span").text
                    lieu_full = lieu_full.split()[-1]
                    lieu_full = lieu_full.replace(")",  "")
                    lieu = lieu_full.replace("(",  "")
                    lieu = lieu.strip()
                except:
                    lieu = "null"
                
                try:
                    price_full = soup.find('li' , {'class' : 'px'}).find("span").text
                    price = price_full.replace('€' , '')
                    price = price.strip()
                except:
                    price = "null"
                print( 'marque: ' + str(marque))
                print('version: ' + str(version)) 
                print('année de fabrication : ' + str(date)) 
                print('kilometrage : ' + str(kil))
                print('energie : ' + str(energie)) 
                print('emission :' + str(emission)) 
                print("consomation :" + str(cons)) 
                print('transmission : ' + str(transmission))
                print('porte :' + str(porte)) 
                print('cv :' + str(cv)) 
                print('place :' + str(place)) 
                print('lieu :' + str(lieu)) 
                print('prix :' + str(price))
                print()
                outf.write(marque + ',' + version + ',' + date + ',' + kil + ',' + energie + ',' + emission + ',' + cons + ',' + transmission + ','+ porte + ',' + cv + ',' + place + ',' + lieu + ',' + price + '\n')
            time.sleep(5)

```
